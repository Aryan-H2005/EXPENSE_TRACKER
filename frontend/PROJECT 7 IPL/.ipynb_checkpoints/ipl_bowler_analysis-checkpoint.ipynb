```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("IPL2025Bowlers.csv")
df.head()
```

```python
top_wkts=df.sort_values(by='WKT',ascending=False).head(10)
print(top_wkts[['Player Name','Team', 'WKT']])
```

```python
plt.figure(figsize=(20,5))
plt.bar(top_wkts['Player Name'], top_wkts['WKT'],color='orange')
plt.xlabel('Player Name')
plt.ylabel('WKT')
plt.title('Top 10 Wicket Takers - IPL 2025')
plt.show()
```

```python
economy=df.sort_values(by='ECO',ascending=True).head(10)
print(economy[['Player Name','ECO']])
```

```python
plt.figure(figsize=(20,8))
plt.bar(economy['Player Name'],economy['ECO'],color='purple')
plt.xlabel('Player Name')
plt.ylabel('Economy')
plt.title('Top 10 Players With Lowest Economy - IPL 2025')
plt.show()
```

```python
plt.figure()
plt.scatter(df['MAT'],df['WKT'],color='teal')
plt.xlabel('Matches')
plt.ylabel('Wickets')
plt.title('Matches v/s Wickets - IPL 2025')
plt.show()
```

```python
df_hauls = df[(df["4W"] > 0) | (df["5W"] > 0)]
df["4W"] = pd.to_numeric(df["4W"], errors="coerce").fillna(0)
df["5W"] = pd.to_numeric(df["5W"], errors="coerce").fillna(0)
```

```python
top_bowlers = df.sort_values("WKT", ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_bowlers["Player Name"], top_bowlers["4W"], label="4W Hauls")
plt.bar(
    top_bowlers["Player Name"],
    top_bowlers["5W"],
    bottom=top_bowlers["4W"],
    label="5W Hauls"
)

plt.xlabel("Player Name")
plt.ylabel("Number of Hauls")
plt.title("4W and 5W Hauls  IPL Bowlers 2025")
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
