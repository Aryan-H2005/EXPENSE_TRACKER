```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("IPL2025Batters.csv")
df.head()
```

```python
top_runs=df.sort_values(by='Runs', ascending=False).head(10)
print(top_runs[['Player Name','Team', 'Runs']])
```

```python
plt.figure(figsize=(20,7))
plt.bar(top_runs['Player Name'], top_runs['Runs'])
plt.xlabel('Player Name')
plt.ylabel('Runs')
plt.title('Top 10 Runs Scorer - IPL 2025')
plt.show()
```

```python
players_per_team=df.groupby("Team")['Player Name'].nunique().sort_values(ascending=False)
print(players_per_team)
```

```python
plt.figure()
players_per_team.plot(kind='bar',color='green')
plt.xlabel('Team')
plt.ylabel('Player Name')
plt.title('Total Players Per Team - IPL 2025')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Runs'],df['SR'], color='purple')
plt.xlabel('Runs')
plt.ylabel('Strike Rate')
plt.title('Runs v/s Strike Rate - IPL 2025')
plt.show()
```

```python
team_runs=df.groupby('Team')['Runs'].sum().sort_values(ascending=False).head(10)
print(team_runs)
```

```python
plt.figure()
team_runs.plot(kind='bar',color='orange')
plt.xlabel('Team')
plt.ylabel('Runs')
plt.title('Team Wise Total Runs - IPL 2025')
plt.show()
```
